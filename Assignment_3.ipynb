```python
url = "https://www.amazon.in/gp/bestsellers/books/"
response = requests.get(url)
data = []
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    for i in range(50):
        div_id = f"p13n-asin-index-{i}"
        div = soup.find("div", {"id": div_id})
        if div:
            a_tag = div.find("a", {"class": "a-link-normal", "tabindex": "-1"})
            if a_tag:
                href = a_tag.get("href")
                data.append({"URL": href})
            else:
                print(f"Specified <a> tag not found for div with id {div_id}.")
        else:
            print(f"Div with id {div_id} not found on the page.")
    df = pd.DataFrame(data)
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
base_url = 'https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page='
num_movies = 1000
movies_data = []
page = 1
while len(movies_data) < num_movies:
    url = base_url + str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    movie_containers = soup.find_all('div', class_='lister-item mode-advanced')
    for container in movie_containers:
        movie_name = container.h3.a.text.strip()
        imdb_rating = container.find('div', class_='ratings-imdb-rating').strong.text.strip()
        description = container.find('div', class_='lister-item-content').find_all('p')[1].text.strip()
        votes = container.find('span', attrs={'name': 'nv'}).text.strip()
        directors = container.find('p', class_='').find_all('a')
        director_names = [director.text for director in directors]
        stars = container.find('p', class_='').find_all('a')[1:]
        star_names = [star.text for star in stars]
        metascore_element = container.find('span', class_='metascore')
        metascore = metascore_element.text.strip() if metascore_element else 'N/A'
        genre = container.find('span', class_='genre').text.strip()
        duration = container.find('span', class_='runtime').text.strip()
        movie_data = {
            'Movie Name': movie_name,
            'IMDb Rating': imdb_rating,
            'Description': description,
            'Votes': votes,
            'Director(s)': ', '.join(director_names),
            'Stars': ', '.join(star_names),
            'Metascore': metascore,
            'Genre': genre,
            'Total Duration': duration
        }
        movies_data.append(movie_data)
        print(movie_data)
        if len(movies_data) == num_movies:
            break
    page += 1
for movie_data in movies_data:
    print('Movie Name:', movie_data['Movie Name'])
    print('IMDb Rating:', movie_data['IMDb Rating'])
    print('Description:', movie_data['Description'])
    print('Votes:', movie_data['Votes'])
    print('Director(s):', movie_data['Director(s)'])
    print('Stars:', movie_data['Stars'])
    print('Metascore:', movie_data['Metascore'])
    print('Genre:', movie_data['Genre'])
    print('Total Duration:', movie_data['Total Duration'])
    print()
```

```python
csv_file = 'movies_data.csv'
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Movie Name', 'IMDb Rating', 'Description', 'Votes', 'Director(s)', 'Stars', 'Metascore', 'Genre', 'Total Duration'])
    writer.writeheader()
    writer.writerows(movies_data)
print('CSV file saved successfully.')
```
